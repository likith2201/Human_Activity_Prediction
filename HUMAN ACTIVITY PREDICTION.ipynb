```python
import seaborn as sns
plt.figure(figsize=(20,10))
sns.heatmap(corr)   # GENERALLY IF WE HAVE SMALL VALUES WE CAN SEE IT ON THIS BUT THIS IS MESSY NOW SO, WE USE PYTHON CODE TO FILTER
```

```python
def corelation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = X_train.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features=corelation(X_train,0.9)
len(set(corr_features))              # LENGTH OF CORR_FEATURES
```

```python
X_train=X_train.drop(corr_features,axis=1)
```

```python
X_test=X_test.drop(corr_features,axis=1)
```
